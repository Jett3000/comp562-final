```python
#Info from https://github.com/spacetelescope/notebooks/blob/master/notebooks/MAST/TESS/beginner_how_to_use_lc/beginner_how_to_use_lc.ipynb

%matplotlib inline
from astropy.io import fits
from astropy.io import ascii
from astropy.table import Table
import scipy.interpolate as interp

import matplotlib.pyplot as plt
import resampy
import math
import numpy as np
import pandas as pd
```

```python
with open('kepler_exo_urls.txt', 'r') as text:
    textfile = []
    for i in range(500):
        textfile.append(text.readline())
        
with open('kepler_non_exo_urls.txt', 'r') as text:
    textfileNon = []
    for i in range(5000):
        textfileNon.append(text.readline())
```

```python
with fits.open(textfileNon[3739], memmap=True) as hdu_list:
    fluxes = hdu_list[1].data['PDCSAP_FLUX']
    fluxes = np.float64(fluxes)
    print(len(fluxes))
```

```python
first = True
num = 1
for url in textfile:
    with fits.open(url, memmap=True) as hdu_list:
        fluxes = hdu_list[1].data['PDCSAP_FLUX']
        fluxes = np.float64(fluxes)
        avg = np.nanmean(fluxes)
        for i in range(len(fluxes)):
            if (math.isnan(fluxes[i])):
                fluxes[i] = avg
        arr2_interp = interp.interp1d(np.arange(fluxes.size), fluxes)
        arr2 = arr2_interp(np.linspace(0, fluxes.size-1, 3197))
        print("Star: " + str(num))
        num += 1
        if (not first):
            fluxes = [2]
            fluxes = np.append(fluxes, arr2)
            newTable.add_row(fluxes)
        else:
            first = False
            newTable = Table()
            newTable['LABEL'] = [2]
            test = 0
            for i in range(len(arr2)):
                test += 1
                newTable['FLUX.' + str(test)] = arr2[test - 1]   
            print(newTable)            
```
